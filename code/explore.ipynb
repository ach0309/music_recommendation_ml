### Univariate Analysis

```python
# check for how much each genre is represented in the dataset
genre_counts = df['genre'].value_counts()


plt.figure(figsize=(10, 6))
genre_counts.plot(kind='bar')
plt.title('Distribution of Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
numerical_cols = ['dating', 'violence', 'world/life', 'night/time', 
                'romantic', 'sadness', 'feelings', 'obscene']  # etc.

df[numerical_cols].hist(figsize=(14, 10), bins=30, color='steelblue', edgecolor='black')
plt.suptitle('Distribution of Lyrical Feature Scores', y=1.02)
plt.tight_layout()
plt.show()
```
